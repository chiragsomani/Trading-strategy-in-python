```python
import fastquant as fq
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
from fastquant import get_crypto_data
## read somewhere on GitHUb, fastquant fetches data from any of the top six exchange on https://www.coingecko.com/en/exchanges
crypto1 = get_crypto_data("BTC/USDT", "2020-12-01", "2022-01-12")
```

```python
## Yahoo Finance fetches data from CoinMarketCap
df = yf.download('BTC-USD', start='2020-12-01')
df.head()
```

```python
crypto2 = df.iloc[1:]
crypto2.head()
## dropping values from 30 Novemeber
```

```python
data = [crypto1['close'], crypto2['Close']]
headers = ['Close1', 'Close2']
new_df = pd.concat(data, axis=1, keys=headers)
new_df['pos_at_1Exchange'] = np.where(new_df['Close1'] > new_df['Close2'], 1, -1) 
new_df['pos_at_2Exchange'] = np.where(new_df['Close1'] < new_df['Close2'], 1, -1)
new_df['return'] = abs(new_df['Close1'] - new_df['Close2'])
new_df.head()
```
